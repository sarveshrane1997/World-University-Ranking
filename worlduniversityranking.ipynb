## operations on null values

### fill null values with mode or mean those columns has null values between 1 % to 40 %

```python
df["ifr rank"]=df["ifr rank"].fillna(df["ifr rank"].mode())
```

```python
df["ifr score"]=df["ifr score"].fillna(df["ifr score"].mean())
```

```python
df["isr score"]=df["isr score"].fillna(df["isr score"].mean())
```

```python
df["isr rank"]=df["isr rank"].fillna(df["isr rank"].mode())
```

### delete column whose null values is more than 40%

```python
df.drop("score scaled",axis=1,inplace=True)
```

### remove rows those have remaining null value(less than 1% columns null value)

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

<b style="color:red">Insights</b>
<p>Now There is no null values in dataset<p>

## columns List regarding their data type

```python
cat_col=df.select_dtypes(include=["O"]).columns
cat_col
```

```python
num_col=df.select_dtypes(include=["int","float"]).columns
num_col
```

<b style="color:red">Insights</b>
<p>there is many columns that mismatch datatype and columns data</p>

### remove the unneccessary + sign from data

```python
def change(a):
    if "+" in a:
        a=a.replace("+","")
        return a
    else:
        return a
```

```python
for i in cat_col:
        df[i]=df[i].apply(change)
```

## change the columns datatypes those mismatch their datatype and their data

```python
def changetype(a):
    for i in a:
        if "score" in i:
            df[i]=df[i].astype("float")
        elif "rank" in i:
            df[i]=df[i].astype(int)
    
```

```python
a=["er rank",'ar rank','er score',"ar score",'fsr rank',"fsr score",'cpf rank',"cpf score",'ifr rank',"ifr score",'isr rank',"isr score",'irn rank',"irn score",'ger rank',"ger score"]
changetype(a)
df["Rank"]=df["Rank"].astype("int")
```

### Q4. which country's most of universities in top 200 rank?

```python
g=df.groupby("location")["institution","Rank"]
b=[0]
c=[0]
for i in df["location"].unique():
    count=0
    for j in g.get_group(i)["institution"]:
        if df[(df["location"]==i)&(df["institution"]==j)]["Rank"].unique()<200:
            count+=1
    if c[0]<count:
        c.pop(0)
        c.append(count)
        b.pop(0)
        b.append(i)
print(b[0])
```

### Q.18 compare between India and Pakistan Universities?

```python
plt.figure(figsize=(14,7))
y=["ar score",'fsr score',"cpf score","ifr score","isr score","irn score","er score","ger score"]
g=df.groupby("location")[["ar score",'fsr score',"cpf score","ifr score","isr score","irn score","er score","ger score"]]
plt.plot(y,g.get_group("India").mean(),'--',label="India",color="Blue")
plt.plot(y,g.get_group("Pakistan").mean(),'--',label="Pakistan",color="green")
plt.legend(loc="best")
plt.xlabel('Score',fontsize=15,color="r")
plt.ylabel('Frequency',fontsize=15,color="r")
plt.xticks(rotation=90)
plt.show()
```

<b style="color:red">Insights</b>
<p>(1)Pakistan highest than Indian in irn and er score<br>
(2)India highest than Pakistan in other score.<br>
